Dataset has many values which are out of bounds, usually in cases where the object is only partially visable. This function should clamp those values into the acceptable range

```python
import os

def clamp(value, min_value=0, max_value=1):
    return max(min_value, min(value, max_value))

def process_labels(label_dir):
    for filename in os.listdir(label_dir):
        if filename.endswith('.txt'):
            filepath = os.path.join(label_dir, filename)
            with open(filepath, 'r') as file:
                lines = file.readlines()
            with open(filepath, 'w') as file:
                for line in lines:
                    parts = line.strip().split()
                    if len(parts) == 5:
                        cls, x, y, w, h = parts
                        x, y, w, h = map(float, [x, y, w, h])
                        x = clamp(x)
                        y = clamp(y)
                        w = clamp(w)
                        h = clamp(h)
                        file.write(f"{cls} {x} {y} {w} {h}\n")
                    else:
                        file.write(line)

# Specify the path to your labels directory
train_dir = 'C:\\Users\\Jack Bowman\\Documents\\Programs\\PytScripts\\CV\\YOLO_video\\dataset\\Dataset military equipment\\images_train'
val_dir = 'C:\\Users\\Jack Bowman\\Documents\\Programs\\PytScripts\\CV\\YOLO_video\\dataset\\Dataset military equipment\\images_val'
process_labels(train_dir)
process_labels(val_dir)
```

```python
from ultralytics import YOLO

# Initialize the model with the pre-trained weights
model = YOLO('yolov9c.pt')

# Path to your dataset configuration file
dataset_yaml = 'C:\\Users\\Jack Bowman\\Documents\\Programs\\PytScripts\\CV\\YOLO_video\\mil_decision.yaml'

# Train the model on your dataset
model.train(data=dataset_yaml, epochs=100, batch=-1)

# Evaluate the model on the validation set
results = model.val(data=dataset_yaml)
print("Results on Validation Set:", results)
model.save('trained_yolov9c_100.pt')

```
